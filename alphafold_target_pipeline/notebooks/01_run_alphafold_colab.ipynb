# 01 — Running AlphaFold (ColabFold)

This notebook documents how AlphaFold2 predictions were generated for  
**Human Dihydrofolate Reductase (DHFR)** using the official ColabFold AlphaFold2 notebook.

---

## Workflow Summary

1. Obtain target protein sequence from UniProt (DHFR, UniProt ID **P00374**).
2. Open the official ColabFold AlphaFold2 notebook in Google Colab:  
   https://colab.research.google.com/github/sokrypton/ColabFold/blob/main/AlphaFold2.ipynb
3. Paste the DHFR sequence into the `query_sequence` input box.
4. Set `jobname = DHFR` and keep other settings as default.
5. Run the notebook via **Runtime → Run all**.
6. Download the ranked `.pdb` model files and store them in this repository under  
   `alphafold_target_pipeline/output/structures/`.

This notebook is for documentation only; the actual AlphaFold2 computation runs in ColabFold.
