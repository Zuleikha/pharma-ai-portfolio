ðŸ“Š Molecular Property Histograms â€” What They Mean
ðŸ§ª Molecular Weight (MW)

Shows how heavy your molecules are.
Interpretation:

200â€“500 Da â†’ ideal range for drug-like molecules

>500 Da â†’ may reduce oral absorption

<200 Da â†’ very small / very flexible

ðŸ’§ LogP (Lipophilicity)

Shows how greasy vs water-loving your molecules are.
Interpretation:

1â€“3 â†’ balanced (ideal for oral drugs)

>5 â†’ too lipophilic (poor solubility & toxicity risk)

<1 â†’ too hydrophilic (poor membrane permeability)

â­ QED (Drug-Likeness)

Overall drug-likeness score on a 0â€“1 scale.
Interpretation:

0.7â€“1.0 â†’ excellent drug-like profile

0.4â€“0.7 â†’ moderate

0â€“0.4 â†’ weak

```python
import matplotlib.pyplot as plt

# Molecular Weight Histogram
plt.figure()
df["MolecularWeight"].hist(bins=10)
plt.title("Molecular Weight Distribution")
plt.xlabel("Molecular Weight")
plt.ylabel("Count")
plt.show()

# LogP Histogram
plt.figure()
df["LogP"].hist(bins=10)
plt.title("LogP Distribution")
plt.xlabel("LogP")
plt.ylabel("Count")
plt.show()

# QED Histogram
plt.figure()
df["QED"].hist(bins=10)
plt.title("QED Score Distribution")
plt.xlabel("QED Score")
plt.ylabel("Count")
plt.show()
```

ðŸ“ˆ Additional Molecular Property Visualizations

Shows how your molecules cluster in â€œdrug-like chemical space.â€
Useful for seeing whether molecules are balanced in size & lipophilicity.

â­ LogP vs QED

Reveals how lipophilicity impacts drug-likeness.
A strong downward trend often indicates too much lipophilicity hurting QED.

ðŸŒ TPSA Distribution

Topological Polar Surface Area (TPSA) relates to permeability:

< 90 Ã…Â² â†’ good for oral absorption

ðŸ”— Correlation Heatmap

Shows relationships between key properties:

ðŸŽ¯ Summary

These plots provide deeper insights into ADME properties and how your molecules behave in chemical space.

```python
import matplotlib.pyplot as plt

# MW vs LogP
plt.figure()
plt.scatter(df["MolecularWeight"], df["LogP"])
plt.title("Chemical Space: MW vs LogP")
plt.xlabel("Molecular Weight")
plt.ylabel("LogP")
plt.show()

# LogP vs QED
plt.figure()
plt.scatter(df["LogP"], df["QED"])
plt.title("LogP vs QED (Drug-Likeness)")
plt.xlabel("LogP")
plt.ylabel("QED Score")
plt.show()

# TPSA Histogram
plt.figure()
df["TPSA"].hist(bins=10)
plt.title("TPSA Distribution")
plt.xlabel("Topological Polar Surface Area (Ã…Â²)")
plt.ylabel("Count")
plt.show()

# Correlation Heatmap
plt.figure()
df_corr = df[["MolecularWeight", "LogP", "QED", "TPSA"]].corr()
plt.imshow(df_corr, cmap="viridis", interpolation="nearest")
plt.colorbar()
plt.title("Correlation Heatmap of Key Properties")
plt.xticks(range(len(df_corr.columns)), df_corr.columns, rotation=45)
plt.yticks(range(len(df_corr.columns)), df_corr.columns)
plt.show()
```
